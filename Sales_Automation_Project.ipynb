```python
import pandas as pd
import numpy as np
import random

from faker import Faker
from datetime import datetime

from openpyxl import load_workbook
from openpyxl.styles import Font, PatternFill
from openpyxl.chart import BarChart, PieChart, Reference

from google.colab import files
```

```python
# Create Faker object to generate realistic fake data

fake = Faker()

# Define possible sales regions

regions = ["East", "West", "South", "Midwest"]

# Define list of sales representatives

reps = [
    "George",
    "Ashley",
    "Josh",
    "Kamil",
    "Sarah",
    "Michael"
]

statuses = ["Closed Won", "Closed Lost"]

sales_data = []

# Generate 500 fake sales transactions

for i in range(1, 501):

    sales_data.append({
        "order_id": 1000 + i,
        "rep_name": random.choice(reps),
        "region": random.choice(regions),
        "sales_amount": random.randint(200, 5000),
        "deal_status": random.choice(statuses),
        "date": fake.date_between(
            start_date='-6M',
            end_date='today'
        )
    })

df = pd.DataFrame(sales_data)

df.head()
```

```python
csv_file = "sales_data.csv"

df.to_csv(csv_file, index=False)

print("CSV file created successfully.")
```

```python
# Sales by Representative

sales_by_rep = (
    df.groupby("rep_name")["sales_amount"]
    .sum()
    .reset_index()
)

# Sales by Region

sales_by_region = (
    df.groupby("region")["sales_amount"]
    .sum()
    .reset_index()
)

# KPI Metrics

total_sales = df["sales_amount"].sum()

average_sale = df["sales_amount"].mean()

total_orders = len(df)

print("Total Sales:", total_sales)

print("Average Sale:", round(average_sale, 2))

print("Total Orders:", total_orders)
```

```python
excel_file = "Sales_Performance_Report.xlsx"

# Create Excel workbook with multiple worksheets

with pd.ExcelWriter(
    excel_file,
    engine="openpyxl"
) as writer:

    df.to_excel(
        writer,
        sheet_name="Raw Data",
        index=False
    )

    sales_by_rep.to_excel(
        writer,
        sheet_name="Sales By Rep",
        index=False
    )

    sales_by_region.to_excel(
        writer,
        sheet_name="Sales By Region",
        index=False
    )

print("Excel workbook created.")
```

```python
wb = load_workbook(excel_file)

# Apply professional formatting to worksheet headers

header_fill = PatternFill(
    start_color="1F4E78",
    end_color="1F4E78",
    fill_type="solid"
)

header_font = Font(
    color="FFFFFF",
    bold=True
)

for sheet in wb.sheetnames:

    ws = wb[sheet]

    for cell in ws[1]:

        cell.fill = header_fill
        cell.font = header_font

    # Auto-size columns

    for column_cells in ws.columns:

        length = max(
            len(str(cell.value))
            if cell.value else 0
            for cell in column_cells
        )

        ws.column_dimensions[
            column_cells[0].column_letter
        ].width = length + 5

wb.save(excel_file)

print("Workbook formatting completed.")
```

```python
wb = load_workbook(excel_file)

# Create Dashboard Sheet

if "Dashboard" in wb.sheetnames:
    del wb["Dashboard"]

ws_dashboard = wb.create_sheet("Dashboard")

# Dashboard Title

ws_dashboard["A1"] = "Sales Performance Dashboard"

ws_dashboard["A1"].font = Font(
    size=16,
    bold=True
)

# KPI Metrics

ws_dashboard["A3"] = "Total Sales"
ws_dashboard["B3"] = total_sales

ws_dashboard["A4"] = "Average Sale"
ws_dashboard["B4"] = round(average_sale, 2)

ws_dashboard["A5"] = "Total Orders"
ws_dashboard["B5"] = total_orders

# =========================
# BAR CHART
# =========================

ws_rep = wb["Sales By Rep"]

bar_chart = BarChart()

bar_chart.title = "Sales By Representative"

bar_chart.y_axis.title = "Sales Amount"

bar_chart.x_axis.title = "Representative"

data = Reference(
    ws_rep,
    min_col=2,
    min_row=1,
    max_row=ws_rep.max_row
)

cats = Reference(
    ws_rep,
    min_col=1,
    min_row=2,
    max_row=ws_rep.max_row
)

bar_chart.add_data(
    data,
    titles_from_data=True
)

bar_chart.set_categories(cats)

ws_dashboard.add_chart(
    bar_chart,
    "D2"
)

# =========================
# PIE CHART
# =========================

ws_region = wb["Sales By Region"]

pie_chart = PieChart()

pie_chart.title = "Sales By Region"

data = Reference(
    ws_region,
    min_col=2,
    min_row=1,
    max_row=ws_region.max_row
)

labels = Reference(
    ws_region,
    min_col=1,
    min_row=2,
    max_row=ws_region.max_row
)

pie_chart.add_data(
    data,
    titles_from_data=True
)

pie_chart.set_categories(labels)

ws_dashboard.add_chart(
    pie_chart,
    "D20"
)

wb.save(excel_file)

print("Dashboard and charts created.")
```
