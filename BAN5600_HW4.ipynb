***Problem 7***

```python
# Sample customer data
customers = [
    {"Customer_Name": "John", "Email": "john@email.com", "Request_Type": "Appointment", "Account_Status": "Active", "Reference_Date": "April 20"},
    {"Customer_Name": "Sarah", "Email": "sarah@email.com", "Request_Type": "Payment", "Account_Status": "Past Due", "Reference_Date": "April 15"},
    {"Customer_Name": "Mike", "Email": "mike@email.com", "Request_Type": "Shipping", "Account_Status": "Active", "Reference_Date": "April 18"},
    {"Customer_Name": "", "Email": "invalid@email.com", "Request_Type": "Shipping", "Account_Status": "Active", "Reference_Date": "April 18"}
]

responses = []

index = 0

# While loop to process records
while index < len(customers):

    customer = customers[index]
    index += 1

    name = customer["Customer_Name"]
    request = customer["Request_Type"]
    status = customer["Account_Status"]
    date = customer["Reference_Date"]

    # Skip invalid records
    if name == "":
        continue

    message = ""

    # Template selection using if statements
    if request == "Appointment":
        message = f"Hi {name}, your appointment is confirmed for {date}. We look forward to seeing you!"

    elif request == "Payment":
        if status == "Past Due":
            message = f"Hi {name}, your payment was due on {date}. Please make a payment as soon as possible."
        else:
            message = f"Hi {name}, this is a reminder that your payment is coming up on {date}."

    elif request == "Shipping":
        message = f"Hi {name}, your order has been shipped and is expected to arrive on {date}."

    else:
        message = f"Hi {name}, thank you for reaching out. We will get back to you shortly."

    responses.append({"Customer": name, "Message": message})

    # Example break condition (optional limit)
    if len(responses) == 10:
        break

# Print results
for r in responses:
    print(f"{r['Customer']}: {r['Message']}")
```

***Problem 8***

```python
# Sample shipment data
shipments = [
    {"Order_ID": 101, "Customer_Name": "John", "Ship_Date": 1, "Expected_Delivery": 5, "Actual_Delivery": 7, "Shipment_Status": "Delivered", "Missing_Documents": False},
    {"Order_ID": 102, "Customer_Name": "Sarah", "Ship_Date": 2, "Expected_Delivery": 6, "Actual_Delivery": 6, "Shipment_Status": "Delivered", "Missing_Documents": False},
    {"Order_ID": 103, "Customer_Name": "Mike", "Ship_Date": 3, "Expected_Delivery": 7, "Actual_Delivery": None, "Shipment_Status": "In Transit", "Missing_Documents": True},
    {"Order_ID": 104, "Customer_Name": "Anna", "Ship_Date": 4, "Expected_Delivery": 8, "Actual_Delivery": 10, "Shipment_Status": "Delivered", "Missing_Documents": False},
    {"Order_ID": 105, "Customer_Name": "Chris", "Ship_Date": 5, "Expected_Delivery": 9, "Actual_Delivery": None, "Shipment_Status": "In Transit", "Missing_Documents": False}
]

delayed_shipments = []
missing_info_shipments = []
exception_report = []

index = 0

# While loop to process shipments
while index < len(shipments):

    shipment = shipments[index]
    index += 1

    order_id = shipment["Order_ID"]
    name = shipment["Customer_Name"]
    expected = shipment["Expected_Delivery"]
    actual = shipment["Actual_Delivery"]
    missing_docs = shipment["Missing_Documents"]

    is_delayed = False
    is_missing = False

    # Check for missing info
    if actual is None or missing_docs == True:
        is_missing = True

    # Check for delay
    if actual is not None:
        if actual > expected:
            is_delayed = True

    # Skip clean records
    if not is_delayed and not is_missing:
        continue

    # Store results
    if is_delayed:
        delayed_shipments.append(shipment)

    if is_missing:
        missing_info_shipments.append(shipment)

    exception_report.append({
        "Order_ID": order_id,
        "Customer": name,
        "Delayed": is_delayed,
        "Missing_Info": is_missing
    })

    # Optional break (limit processing)
    if len(exception_report) == 10:
        break

# Output results
print("Delayed Shipments:")
for d in delayed_shipments:
    print(d)

print("\nMissing Info Shipments:")
for m in missing_info_shipments:
    print(m)

print("\nFinal Exception Report:")
for e in exception_report:
    print(e)

# Count summary
print("\nSummary:")
print("Total Delayed:", len(delayed_shipments))
print("Total Missing Info:", len(missing_info_shipments))
```

***Problem 9***

```python
# Sample application data
applications = [
    {"Applicant_Name": "John", "Email": "john@email.com", "Phone": "1234567890", "Date_of_Birth": 1995, "Application_Status": "Pending"},
    {"Applicant_Name": "Sarah", "Email": "sarahemail.com", "Phone": "12345", "Date_of_Birth": 2008, "Application_Status": "Pending"},
    {"Applicant_Name": "Mike", "Email": "", "Phone": "9876543210", "Date_of_Birth": 1985, "Application_Status": "Approved"},
    {"Applicant_Name": "Anna", "Email": "anna@email.com", "Phone": "abcdefghij", "Date_of_Birth": 1975, "Application_Status": "Pending"},
    {"Applicant_Name": "", "Email": "test@email.com", "Phone": "1234567890", "Date_of_Birth": 1990, "Application_Status": "Pending"}
]

missing_fields = []
invalid_formats = []
flagged_records = []

index = 0

# While loop to process applications
while index < len(applications):

    app = applications[index]
    index += 1

    name = app["Applicant_Name"]
    email = app["Email"]
    phone = app["Phone"]
    dob = app["Date_of_Birth"]

    is_missing = False
    is_invalid = False

    # Check for missing fields using a for loop
    required_fields = [name, email, phone, dob]

    for field in required_fields:
        if field == "" or field is None:
            is_missing = True

    # Email validation
    if "@" not in email or email == "":
        is_invalid = True

    # Phone validation
    if not phone.isdigit() or len(phone) != 10:
        is_invalid = True

    # Age validation
    current_year = 2026
    age = current_year - dob

    if age < 18:
        is_invalid = True

    # Skip clean records
    if not is_missing and not is_invalid:
        continue

    # Store results
    if is_missing:
        missing_fields.append(app)

    if is_invalid:
        invalid_formats.append(app)

    flagged_records.append({
        "Applicant": name,
        "Missing": is_missing,
        "Invalid": is_invalid
    })

    # Optional break
    if len(flagged_records) == 10:
        break

# Output results
print("Missing Field Records:")
for m in missing_fields:
    print(m)

print("\nInvalid Format Records:")
for i in invalid_formats:
    print(i)

print("\nValidation Report:")
for f in flagged_records:
    print(f)

print("\nSummary:")
print("Total Missing:", len(missing_fields))
print("Total Invalid:", len(invalid_formats))
```

```python
# Sample task data
tasks = [
    {"Task_ID": 1, "Task_Name": "Submit Report", "Assigned_To": "George", "Due_Date": 10, "Priority": "High", "Status": "Open"},
    {"Task_ID": 2, "Task_Name": "Email Client", "Assigned_To": "Sarah", "Due_Date": 15, "Priority": "Medium", "Status": "In Progress"},
    {"Task_ID": 3, "Task_Name": "Update Dashboard", "Assigned_To": "Mike", "Due_Date": 5, "Priority": "High", "Status": "Open"},
    {"Task_ID": 4, "Task_Name": "Team Meeting", "Assigned_To": "Anna", "Due_Date": 18, "Priority": "Low", "Status": "Completed"},
    {"Task_ID": 5, "Task_Name": "Code Review", "Assigned_To": "Chris", "Due_Date": 12, "Priority": "High", "Status": "Open"}
]

current_day = 10

overdue_tasks = []
upcoming_tasks = []
priority_groups = {"High": [], "Medium": [], "Low": []}
reminder_report = []

index = 0

# While loop to process tasks
while index < len(tasks):

    task = tasks[index]
    index += 1

    due = task["Due_Date"]
    status = task["Status"]
    priority = task["Priority"]

    is_overdue = False
    is_upcoming = False

    # Skip completed tasks
    if status == "Completed":
        continue

    # Check overdue
    if due < current_day:
        is_overdue = True

    # Check upcoming (within 7 days)
    elif due >= current_day and due <= current_day + 7:
        is_upcoming = True

    # Store results
    if is_overdue:
        overdue_tasks.append(task)

    if is_upcoming:
        upcoming_tasks.append(task)

    # Group by priority
    if priority in priority_groups:
        priority_groups[priority].append(task)

    reminder_report.append({
        "Task": task["Task_Name"],
        "Assigned_To": task["Assigned_To"],
        "Overdue": is_overdue,
        "Upcoming": is_upcoming,
        "Priority": priority
    })

    # Optional break
    if len(reminder_report) == 10:
        break

# Output results
print("Overdue Tasks:")
for o in overdue_tasks:
    print(o)

print("\nUpcoming Tasks (Next 7 Days):")
for u in upcoming_tasks:
    print(u)

print("\nTasks Grouped by Priority:")
for p in priority_groups:
    print(p, ":", priority_groups[p])

print("\nFinal Reminder Report:")
for r in reminder_report:
    print(r)

# Summary
print("\nSummary:")
print("Total Overdue:", len(overdue_tasks))
print("Total Upcoming:", len(upcoming_tasks))
```
