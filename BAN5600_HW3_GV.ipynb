```python
# QUESTION 1 - While Loop: Odd numbers between 150 and 1000

print("=" * 55)
print("QUESTION 1 - While Loop")
print("=" * 55)

# Set everything up before the loop starts
count = 0          # Tracks how many odd numbers we come across
total_sum = 0      # Running total of all the odd numbers
multiplication = 1 # Start at 1 since multiplying by 0 would kill everything
sum_of_squares = 0 # Keeps track of each odd number squared and added together
num = 150          # Starting point for the loop

while num <= 1000:              # Loop through numbers from 150 up to 1000
    if num % 2 != 0:            # If it's not divisible by 2, it's odd
        count += 1              # Increase count when we find an odd number
        total_sum += num        # Add it to the running sum
        multiplication *= num   # Multiply it into the running product
        sum_of_squares += num ** 2  # Square it and add to total
    num += 1                    # Move to the next number every time

print(f"a) Count of odd numbers     : {count}")
print(f"b) Sum of odd numbers       : {total_sum}")
print(f"c) Multiplication of odds   : {multiplication}")
print(f"d) Sum of squares of odds   : {sum_of_squares}")


# QUESTION 2 - For Loop: Odd numbers between 150 and 1000

print("\n" + "=" * 55)
print("QUESTION 2 - For Loop")
print("=" * 55)

# Reset everything again for the new loop
count = 0
total_sum = 0
multiplication = 1
sum_of_squares = 0

# range goes up to 1001 so 1000 is included
for num in range(150, 1001):
    if num % 2 != 0:
        count += 1
        total_sum += num
        multiplication *= num
        sum_of_squares += num ** 2
# The loop will handle so no need to manually increment

print(f"a) Count of odd numbers     : {count}")
print(f"b) Sum of odd numbers       : {total_sum}")
print(f"c) Multiplication of odds   : {multiplication}")
print(f"d) Sum of squares of odds   : {sum_of_squares}")

# QUESTION 3 - Sum from 1 to M using while loop

print("\n" + "=" * 55)
print("QUESTION 3 - Sum from 1 to M")
print("=" * 55)

M = int(input("Enter a value for M: "))  # Get M from the user

s = 0   # Running sum starts at 0
k = 1   # Start counting from 1

while k <= M:   # Keep going until we hit M
    s = s + k   # Add current number to the total
    k = k + 1   # Move to the next number

print(f"Sum from 1 to {M} = {s}")

# QUESTION 4 - Read numbers until "done", print total/count/avg

print("\n" + "=" * 55)
print("QUESTION 4 - Read numbers until 'done'")
print("=" * 55)

total = 0   # Adds up all valid inputs
count = 0   # Counts how many valid numbers we get

while True:                         # Infinite loop until we break out
    entry = input("Enter a number: ")
    if entry == "done":             # Stop when user types "done"
        break
    try:
        number = float(entry)       # Try converting input to a number
        total += number             # Add it to total if it works
        count += 1                  # Increase count
    except ValueError:              # If conversion fails
        print("Invalid input")      # Let user know and keep going

if count > 0:                       # Avoid dividing by 0
    average = total / count
    print(total)
    print(count)
    print(average)
else:
    print("No valid numbers were entered.")

# QUESTION 5 - Read numbers until "done", print max and min

print("\n" + "=" * 55)
print("QUESTION 5 - Read numbers until 'done', print max & min")
print("=" * 55)

# Start with None so we can properly set the first value
max_val = None
min_val = None

while True:
    entry = input("Enter a number: ")
    if entry == "done":
        break
    try:
        number = float(entry)
        # First valid number becomes both max and min automatically
        if max_val is None or number > max_val:
            max_val = number
        if min_val is None or number < min_val:
            min_val = number
    except ValueError:
        print("Invalid input")

if max_val is not None:             # Only print if we actually got numbers
    print(f"Maximum: {max_val}")
    print(f"Minimum: {min_val}")
else:
    print("No valid numbers were entered.")

# QUESTION 6 - What is printed?

print("\n" + "=" * 55)
print("QUESTION 6 - Output of: for z in [2, 4, 7, 9]: print(z - 1)")
print("=" * 55)

# Loop goes through each number in the list
# Each time, we subtract 1 and print it
for z in [2, 4, 7, 9]:
    print(z - 1)

# QUESTION 7 - What is printed?

print("\n" + "=" * 55)
print("QUESTION 7 - Output of the n = n + x loop")
print("=" * 55)

# print runs inside the loop, so we see output each time

n = 3               # Starting value
for x in [2, 5, 8]:
    n = n + x       # Add current value to n
    print(n)        # Print after each update
```

```python
# Replicating Code

from google.colab import auth
import gspread
from google.auth import default

auth.authenticate_user()
creds, _ = default()

gc = gspread.authorize(creds)

print("Successfully authenticated")
```

```python
sh = gc.create('Baseball Player Report')
worksheet = sh.get_worksheet(0)

headers = ["Player Name", "Team", "Batting Avg", "Home Runs", "RBIs"]
worksheet.append_row(headers)
```

```python
player_data = [
    ["Aaron Judge", "NYY", 0.311, 37, 85],
    ["Mookie Betts", "LAD", 0.298, 29, 72],
    ["Jose Ramirez", "CLE", 0.285, 24, 80],
    ["Juan Soto", "NYM", 0.305, 31, 78]
]

print("Writing data to Google Sheets...")

for row in player_data:
  worksheet.append_row(row)

print(f"Done! Check your sheet here: {sh.url}")
```
