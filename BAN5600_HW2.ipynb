**Problem 1:**


```python
import random  # Import random module to gen random numbers
import math    # Import math module to use functions

# Generate two float numbers between 0 and 1
u = random.random()
v = random.random()

# Apply the formula to transform random numbers
# into a standard Gaussian random variable Z
# Formula: Z = sin(2 * pi * v) * sqrt(-2 * ln(u))
Z = math.sin(2 * math.pi * v) * math.sqrt(-2 * math.log(u))

# Display the values of u, v, and the resulting Gaussian random variable Z
print(f"u = {u}")
print(f"v = {v}")
print(f"Z = {Z}")
```

**Problem 2:**

```python
# Use the input() to prompt user
# Ask the user to enter two integer values
a = int(input("Enter first integer: "))
b = int(input("Enter second integer: "))

# Compare the two values and print them in ascending order
if a <= b:
    # if a is already less than or equal to b, print as is
    print(a, b)
else:
    # if b is smaller, so print b first, then a
    print(b, a)
```

```python
# In Jupyter, we can't use sys.argv for input
# Instead, we use input() to prompt the user directly

# Ask the user to enter three float values
x = float(input("Enter first value: "))
y = float(input("Enter second value: "))
z = float(input("Enter third value: "))

# Check if values are strictly ascending (x < y < z)
# OR strictly descending (x > y > z)
if (x < y < z) or (x > y > z):
    print("True")   # Values are in strict order
else:
    print("False")  # Values are not in strict ascending/descending order
```

**Problem 4:**

```python
# Prompt the user to enter an earthquake magnitude as a decimal number
magnitude = float(input("Enter earthquake magnitude: "))

# Use if-elif-else ladder to match the magnitude to its descriptor
# Each condition checks the upper boundary of each range
if magnitude < 2.0:
    descriptor = "Micro"          # Less than 2.0
elif magnitude < 3.0:
    descriptor = "Very Minor"     # 2.0 to less than 3.0
elif magnitude < 4.0:
    descriptor = "Minor"          # 3.0 to less than 4.0
elif magnitude < 5.0:
    descriptor = "Light"          # 4.0 to less than 5.0
elif magnitude < 6.0:
    descriptor = "Moderate"       # 5.0 to less than 6.0
elif magnitude < 7.0:
    descriptor = "Strong"         # 6.0 to less than 7.0
elif magnitude < 8.0:
    descriptor = "Major"          # 7.0 to less than 8.0
elif magnitude < 10.0:
    descriptor = "Great"          # 8.0 to less than 10.0
else:
    descriptor = "Meteoric"       # 10.0 or greater

# Display a meaningful message with the magnitude and its descriptor
print(f"A magnitude {magnitude} earthquake is considered to be a {descriptor} earthquake.")
```

**Problem 6:**

```python
s = 'I ate it.'         # Define a string variable s

print(s.upper())        # .upper() converts ALL characters to uppercase
                        # Output: I ATE IT.

print(s.count('t'))     # .count('t') counts how many times lowercase 't' appears
                        # 't' appears in 'ate' and 'it' → count is 2
                        # Output: 2

w = s.split()           # .split() with no arguments splits on any whitespace
                        # Breaks the string into a list of words
print(w)                # Output: ['I', 'ate', 'it.']
                        # Note: 'it.' keeps its period since split only removes spaces
```
